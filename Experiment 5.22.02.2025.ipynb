# Assuming a set of documents that need to be classified, use the navie Bayesian Classifier model to perform this task.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
msg = pd.read_csv('document.csv', names=['message', 'label'], encoding='latin-1')
print("Total Instances of Dataset: ", msg.shape[0])
msg['labelnum'] = msg['label'].map({'pos': 1, 'neg' : 0})
X = msg['message']
Y = msg['labelnum']
print(msg)
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y)
```

```python
vectorizer = CountVectorizer()
Xtrain_dm = vectorizer.fit_transform(Xtrain)
Xtest_dm = vectorizer.transform(Xtest)
```

```python
Xtrain_dm
```

```python
clf = MultinomialNB()
clf.fit(Xtrain_dm, Ytrain)
```

```python
pred = clf.predict(Xtest_dm)
print('Accuracy: \n', accuracy_score(Ytest, pred))
print('Confusion Matrix:\n', confusion_matrix(Ytest, pred))
```
