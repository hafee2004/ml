# Experiment 2:Data Analysis & Visualization on diwali sales dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('dataset.csv')
```

```python
df
```

```python
x=df['Age'].head(10)
y=df['DailyRate'].head(10)
```

```python
plt.scatter(x,y,color='Red')
plt.xlabel('Age')
plt.ylabel('DailyRate')
plt.title('Age Vs Daily Rate Scatter Plot')
```

```python
c= np.array(['red', 'green', 'yellow', 'magenta', 'cyan', 'blue', 'black', 'orange', 'purple', 'brown'])
x1=df['Age'].head(10)
y1=df['DailyRate'].head(10)

plt.scatter(x1,y1,color=c,s=200,marker='*') 
plt.xlabel('Age')
plt.ylabel('DailyRate')
plt.title('Age Vs Daily Rate Scatter Plot')
plt.colorbar()
```

```python
plt.bar(x,y, color = 'green')
plt.xlabel('Age')
plt.ylabel('DailyRate')
plt.title('Age Vs Daily Rate Scatter Plot')
```

```python
plt.bar(x1,y1, color = c)
plt.xlabel('Age')
plt.ylabel('DailyRate')
plt.title('Age Vs Daily Rate Scatter Plot')
```
